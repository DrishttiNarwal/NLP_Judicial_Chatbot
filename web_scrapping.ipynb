```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.action_chains import ActionChains
import time
import csv
```

```python
# Initialize the WebDriver
service = Service(r"C:\Users\chris\Downloads\chromedriver-win64\chromedriver-win64\chromedriver.exe")  # Replace with your chromedriver path
driver = webdriver.Chrome(service=service)

# Open the webpage (replace with actual webpage containing the menu structure)
driver.get("https://doj.gov.in")  # Replace with your URL
```

```python
# Maximize window to ensure all elements are visible
driver.maximize_window()
```

```python
# Find the main navigation bar by its ID
main_menu = driver.find_element(By.ID, 'menu-header-primary')

# # Find all <a> tags within the menu and sub-menu
menu_items = main_menu.find_elements(By.TAG_NAME, 'a')
```

```python
def extract_paragraphs_and_headings():
    # Open CSV file once before the loop
    with open('menu_links_content.csv', 'w', newline='', encoding='utf-8') as file:
        writer = csv.writer(file)
        writer.writerow(['links', 'content'])  # Write header

        # Re-locate main menu and its items before starting the loop
        main_menu = driver.find_element(By.ID, 'menu-header-primary')
        menu_items = main_menu.find_elements(By.TAG_NAME, 'a')

        # Iterate through each link
        for index in range(len(menu_items)):
            try:
                # Re-locate elements for every loop iteration to avoid stale element exception
                main_menu = driver.find_element(By.ID, 'menu-header-primary')
                menu_items = main_menu.find_elements(By.TAG_NAME, 'a')
                
                item = menu_items[index]
                link_text = item.text
                link_url = item.get_attribute('href')
                print(f"Processing: {link_text} - {link_url}")

                # Skip if no href (i.e., no valid link)
                if not link_url:
                    continue

                # Open the link in the same browser tab
                driver.get(link_url)
                time.sleep(2)  # Wait for the page to load

                # Extract paragraphs and headings
                paragraphs = driver.find_elements(By.TAG_NAME, 'p')
                headings = []
                for i in range(1, 7):  # Loop through <h1> to <h6>
                    headings.extend(driver.find_elements(By.TAG_NAME, f'h{i}'))

                # Collect the text content
                page_content = ""
                page_content += "\n".join([p.text for p in paragraphs])  # Extract paragraphs
                page_content += "\n".join([h.text for h in headings])  # Extract headings

                # Write the URL and extracted content to CSV
                writer.writerow([link_url, page_content])

                # Navigate back to the main page to continue processing
                driver.get("https://doj.gov.in")
                time.sleep(2)  # Wait for the page to load before finding the menu again

            except Exception as e:
                print(f"Error processing item {index}: {str(e)}")

# Run the extraction function
extract_paragraphs_and_headings()

# Close the driver after completion
driver.quit()
```
